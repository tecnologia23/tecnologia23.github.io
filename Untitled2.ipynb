```python
# hagamos un programa que me calcule el factorial al número y que el, usuario ingrese el numero.

def calcular_factorial(numero):
  """Calcula el factorial de un número entero no negativo."""
  if numero < 0:
    return "No se puede calcular el factorial de un número negativo."
  elif numero == 0:
    return 1
  else:
    factorial = 1
    for i in range(1, numero + 1):
      factorial *= i
    return factorial

# Solicitar al usuario que ingrese un número
try:
  num = int(input("Ingresa un número entero no negativo para calcular su factorial: "))
  resultado = calcular_factorial(num)
  print(f"El factorial de {num} es: {resultado}")
except ValueError:
  print("Entrada inválida. Por favor, ingresa un número entero.")
```

```python
def calcular_factorial_recursivo(numero):
  """Calcula el factorial de un número entero no negativo utilizando recursión."""
  if numero < 0:
    return "No se puede calcular el factorial de un número negativo."
  elif numero == 0:
    return 1
  else:
    return numero * calcular_factorial_recursivo(numero - 1)

try:
  num = int(input("Ingresa un número entero no negativo para calcular su factorial: "))
  resultado = calcular_factorial_recursivo(num)
  print(f"El factorial de {num} es: {resultado}")
except ValueError:
  print("Entrada inválida. Por favor, ingresa un número entero.")
```

```python
import random

def mostrar_menu():
    """Muestra las opciones disponibles para el jugador"""
    print("\n" + "="*40)
    print("🎮 PIEDRA, PAPEL Y TIJERA 🎮")
    print("="*40)
    print("1. 🪨 Piedra")
    print("2. 📄 Papel")
    print("3. ✂️  Tijera")
    print("4. 📊 Ver estadísticas")
    print("5. 🚪 Salir")
    print("="*40)

def obtener_eleccion_jugador():
    """Obtiene y valida la elección del jugador"""
    while True:
        try:
            eleccion = int(input("Elige tu opción (1-5): "))
            if eleccion in [1, 2, 3, 4, 5]:
                return eleccion
            else:
                print("❌ Por favor, elige un número del 1 al 5")
        except ValueError:
            print("❌ Por favor, ingresa un número válido")

def obtener_eleccion_computadora():
    """Genera una elección aleatoria para la computadora"""
    return random.randint(1, 3)

def convertir_a_texto(eleccion):
    """Convierte el número de elección a texto"""
    opciones = {1: "Piedra", 2: "Papel", 3: "Tijera"}
    return opciones[eleccion]

def convertir_a_emoji(eleccion):
    """Convierte el número de elección a emoji"""
    emojis = {1: "🪨", 2: "📄", 3: "✂️"}
    return emojis[eleccion]

def determinar_ganador(jugador, computadora):
    """Determina el ganador de la partida"""
    if jugador == computadora:
        return "empate"
    elif (jugador == 1 and computadora == 3) or \
         (jugador == 2 and computadora == 1) or \
         (jugador == 3 and computadora == 2):
        return "jugador"
    else:
        return "computadora"

def mostrar_resultado(jugador, computadora, resultado):
    """Muestra el resultado de la partida"""
    print(f"\n🎯 RESULTADO:")
    print(f"Tú elegiste: {convertir_a_emoji(jugador)} {convertir_a_texto(jugador)}")
    print(f"Computadora eligió: {convertir_a_emoji(computadora)} {convertir_a_texto(computadora)}")

    if resultado == "empate":
        print("🤝 ¡Es un empate!")
    elif resultado == "jugador":
        print("🎉 ¡Ganaste!")
    else:
        print("💻 ¡La computadora ganó!")

def mostrar_estadisticas(stats):
    """Muestra las estadísticas del juego"""
    total = stats["jugador"] + stats["computadora"] + stats["empates"]
    if total == 0:
        print("\n📊 No hay partidas jugadas aún")
        return

    print(f"\n📊 ESTADÍSTICAS:")
    print(f"Total de partidas: {total}")
    print(f"Victorias del jugador: {stats['jugador']} ({stats['jugador']/total*100:.1f}%)")
    print(f"Victorias de la computadora: {stats['computadora']} ({stats['computadora']/total*100:.1f}%)")
    print(f"Empates: {stats['empates']} ({stats['empates']/total*100:.1f}%)")

def jugar():
    """Función principal del juego"""
    print("¡Bienvenido al juego de Piedra, Papel y Tijera!")

    # Estadísticas del juego
    estadisticas = {"jugador": 0, "computadora": 0, "empates": 0}

    while True:
        mostrar_menu()
        eleccion_jugador = obtener_eleccion_jugador()

        if eleccion_jugador == 5:  # Salir
            print("\n👋 ¡Gracias por jugar!")
            mostrar_estadisticas(estadisticas)
            break
        elif eleccion_jugador == 4:  # Ver estadísticas
            mostrar_estadisticas(estadisticas)
            continue

        # Jugar una partida
        eleccion_computadora = obtener_eleccion_computadora()
        resultado = determinar_ganador(eleccion_jugador, eleccion_computadora)

        mostrar_resultado(eleccion_jugador, eleccion_computadora, resultado)

        # Actualizar estadísticas
        if resultado == "jugador":
            estadisticas["jugador"] += 1
        elif resultado == "computadora":
            estadisticas["computadora"] += 1
        else:
            estadisticas["empates"] += 1

        # Preguntar si quiere seguir jugando
        while True:
            continuar = input("\n¿Quieres jugar otra vez? (s/n): ").lower().strip()
            if continuar in ['s', 'si', 'sí', 'y', 'yes']:
                break
            elif continuar in ['n', 'no']:
                print("\n👋 ¡Gracias por jugar!")
                mostrar_estadisticas(estadisticas)
                return
            else:
                print("❌ Por favor, responde 's' para sí o 'n' para no")

# Ejecutar el juego
if __name__ == "__main__":
    jugar()
```

```python
import random

# Opciones del juego
opciones = ["piedra", "papel", "tijera"]

print("¡Bienvenido al juego de Piedra, Papel y Tijera!")
print("Opciones: piedra, papel, tijera")

while True:
    # Elección del jugador
    jugador = input("\nTu elección: ").lower()

    # Verificar si quiere salir
    if jugador == "salir":
        print("¡Gracias por jugar!")
        break

    # Verificar si la elección es válida
    if jugador not in opciones:
        print("Opción no válida. Intenta de nuevo.")
        continue

    # Elección de la computadora
    computadora = random.choice(opciones)

    print(f"Tú: {jugador}")
    print(f"Computadora: {computadora}")

    # Determinar el ganador
    if jugador == computadora:
        print("¡Empate!")
    elif (jugador == "piedra" and computadora == "tijera") or \
         (jugador == "papel" and computadora == "piedra") or \
         (jugador == "tijera" and computadora == "papel"):
        print("¡Ganaste!")
    else:
        print("¡Perdiste!")

    # Preguntar si quiere continuar
    continuar = input("\n¿Otra partida? (s/n): ").lower()
    if continuar != "s":
        print("¡Gracias por jugar!")
        break
```
